```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
pixels = data.iloc[:,1:784].values
labels = data.iloc[:,0:1].values
```

```python
pixels
```

```python
pixels = pixels.tolist()
labels = labels.tolist()

training = []
output_list = [0] * 10

for i in range(len(labels)):
    output_row = list(output_list)
    output_row[labels[i][0]] = 1
    training.append([pixels[i], output_row])
    
```

```python
random.shuffle(training)
training = np.array(training)

train_x = list(training[:,0])
train_y = list(training[:,1])

train_x = np.array(train_x)
train_y = np.array(train_y)

print("Train_x",train_x)
print("Train_y",train_y)
```

```python
model = Sequential()
model.add(Dense(100, input_dim=len(train_x[0]), activation='relu'))
model.add(Dense(len(train_y[0]), activation='softmax'))


model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])

model.fit(train_x, train_y, batch_size=256, epochs=150)
```

```python
data2= pd.read_csv("test.csv")
data2.head()
```

```python
teste = data2.iloc[:,0:783].values
```

```python
prediction = model.predict(teste, batch_size=1, verbose=0)
```
